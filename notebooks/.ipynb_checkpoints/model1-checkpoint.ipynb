```python
#To implement the gaussian model over the training data

import os
import cv2
import numpy as np
import random
import os, numpy, PIL
from PIL import Image
import matplotlib .pyplot as plt
import matplotlib.image as mpimg

train_pos_path = '../dataset/train_pos/'
test_pos_path = '../dataset/test_pos/'
train_neg_path = '../dataset/train_neg/'
test_neg_path = '../dataset/test_neg/'

X1 = np.array([np.zeros(10800)]) #This is the feature vector for all positive images
X0 = np.array([np.zeros(10800)]) #This is the feature vector for all negative images


#First finding u1, c1 for all the positives
for filename in os.listdir(train_pos_path):
    imgpath = train_pos_path + filename
    img = cv2.imread(imgpath,-1)
    #print(img.shape)
    x = np.asarray(img)
    x = np.reshape(x,(1,10800))
    #print(x.shape)
    X1 = np.dstack([X1,x]) 
print(X1.shape)

for filename in os.listdir(train_neg_path):
    imgpath = train_neg_path + filename
    img = cv2.imread(imgpath,-1)
    #print(img.shape)
    x = np.asarray(img)
    x = np.reshape(x,(1,10800))
    #print(x.shape)
    X0 = np.dstack([X0,x])
print(X0.shape)

Y1 = np.array([np.ones(1000)])
print(Y1.shape)

Y0 = np.array([np.zeros(1000)])
print(Y0.shape)
```

```python
#Deleting the first row of the feature vector

X0 = np.delete(X0, 0, axis=2)
print(X0.shape)
```

```python
#Reshaping the array into
X0 = np.reshape(X0,(1000,10800))
print(X0.shape)
```

```python
#Reshaping the other arrays

X1 = np.delete(X1, 0, axis=2)
print(X0.shape)

X1 = np.reshape(X1,(1000,10800))
print(X1.shape)
```

```python
#Finding the mean of all the images

# Access all JPG files in directory
allfiles=os.listdir(train_pos_path)
#print(allfiles)

imlist=[filename for filename in allfiles if  filename[-4:] in [".jpg",".JPG"]]

# Assuming all images are the same size, get dimensions of first image
w,h=Image.open(train_pos_path+imlist[0]).size
N=len(imlist)

# Create a numpy array of floats to store the average (assume RGB images)
arr=numpy.zeros((60,60,3),numpy.float)

# Build up average pixel intensities, casting each image as an array of floats
for im in imlist:
    imarr=numpy.array(Image.open(train_pos_path+im),dtype=numpy.float)
    arr=arr+imarr/N

# Round values in array and cast as 8-bit integer
arr=numpy.array(numpy.round(arr),dtype=numpy.uint8)

# Generate, save and preview final image
meanImg1=Image.fromarray(arr,mode="RGB")

############################################################################################################ Modularize this
# Find the mean images of the negative images

# Access all JPG files in directory
allfiles=os.listdir(train_neg_path)
#print(allfiles)

imlist=[filename for filename in allfiles if  filename[-4:] in [".jpg",".JPG"]]

# Assuming all images are the same size, get dimensions of first image
w,h=Image.open(train_neg_path+imlist[0]).size
N=len(imlist)

# Create a numpy array of floats to store the average (assume RGB images)
arr=numpy.zeros((60,60,3),numpy.float)

# Build up average pixel intensities, casting each image as an array of floats
for im in imlist:
    imarr=numpy.array(Image.open(train_neg_path+im),dtype=numpy.float)
    arr=arr+imarr/N

# Round values in array and cast as 8-bit integer
arr=numpy.array(numpy.round(arr),dtype=numpy.uint8)

# Generate, save and preview final image
meanImg0=Image.fromarray(arr,mode="RGB")
```

```python
#Finding the covariance of all the images

covNeg = np.cov(X0.reshape(10800,1000))
covPos = np.cov(X1.reshape(10800,1000))

#print(covPos)
print(covPos.shape)
print(covNeg.shape)

#Displaying the covariance as images

# Round values in array and cast as 8-bit integer
arrNeg=numpy.array(numpy.round(covNeg),dtype=numpy.uint8)
arrPos=numpy.array(numpy.round(covPos),dtype=numpy.uint8)

print(np.diagonal(arrNeg))
print(np.diagonal(arrPos))

arrNeg = np.reshape(np.diagonal(arrNeg),(60,60,3))
arrPos = np.reshape(np.diagonal(arrPos),(60,60,3))

# Generate, save and preview final image
covImg0=Image.fromarray(arrNeg,mode="RGB")
covImg1=Image.fromarray(arrPos,mode="RGB")
```

```python
#Displaying the covariance and the mean of the positive and the negative images

%matplotlib inline

f, axarr = plt.subplots(2,2)

axarr[0,0].imshow(covImg0)
axarr[0,1].imshow(covImg1)
axarr[1,0].imshow(meanImg0)
axarr[1,1].imshow(meanImg1)
```

```python
#Fitting the gaussian model

from scipy.stats import multivariate_normal
from sklearn.decomposition import PCA
from sklearn import preprocessing

#First concatentae the total data
Xfeatures = np.dstack([X1,X0])
Xfeatures = np.reshape(Xfeatures,(2000,10800))
#print(Xfeatures.shape)

#Calculating PCA, reducing the dimensionality

scaler = preprocessing.StandardScaler().fit(Xfeatures)
mean = scaler.mean_
#print(mean)

Xfeatures = scaler.transform(Xfeatures)
#print(Xfeatures.shape)
#print(Xfeatures)

pca = PCA(n_components=50)
#print(pca.explained_variance_ratio)
Xnew = pca.fit_transform(Xfeatures)
#print(Xnew.shape)

#Calculating the mean and the covariance of the face and non face images

mean1 = np.mean(Xnew[:1000],axis=0)
#print(mean1.shape)

cov1 = np.cov(Xnew[:1000].reshape(50,1000))
#print(cov1.shape)

mean2 = np.mean(Xnew[1000:],axis=0)
#print(mean2.shape)

cov2 = np.cov(Xnew[1000:].reshape(50,1000))
#print(cov2.shape)

model1 = multivariate_normal(mean1,cov1) #-- Singular Matrix error when the features is 10800
model2 = multivariate_normal(mean2,cov2)
```

```python
#Run the model on the test images
paths = [test_pos_path,test_neg_path]
threshold = 0.5
TNcount=TPcount=FPcount=FNcount=0

for p in paths:
    for filename in os.listdir(p):
        imgpath = p + filename
        img = cv2.imread(imgpath,-1)
        x = np.asarray(img)
        x = np.reshape(x,(1,10800))
        scaled = scaler.transform(x)
        xtest = pca.transform(scaled)
        
        #Applying Bayes Rule to the pdf's
        res1 = model1.pdf(xtest)/(model1.pdf(xtest) + model2.pdf(xtest))
        #print(res1)
        res2 = model2.pdf(xtest)/(model1.pdf(xtest) + model2.pdf(xtest))
        #print(res2)

        probPos = res1/(res1+res2)
        probNeg = res2/(res1+res2)
        
        if probPos > threshold:
            TPcount = TPcount + 1
        if probPos < threshold:
            FNcount = FNcount + 1
        if probNeg > threshold:
            TNcount = TNcount + 1
        if probNeg < threshold:
            FPcount = FPcount + 1
            
FPR = float(FPcount/100)
TPR = float(TPcount/100)
MR = float(FPcount + FNcount/200)

print("True Positives Count",TPcount)
print("True Negatives Count",TNcount)
print("False Positives Count",FPcount)
print("False Negatives Count",FNcount)
print('False Positive Rate', FPR)
print('True Positive Rate', TPR)
print('Misclassification Rate', MR)
```
